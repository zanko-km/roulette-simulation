<h2>Import Libraries</h2>

```python
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
import random
from scipy.stats import norm
```

<h2>Import Libraries</h2>

<b>I used random choice function in random library for the function where the chance of selecting 1 is `18/38` (win) and the chance of selecting -1 is `20/38` (loss).</b>

```python
def simulate_N(N):
    money = np.random.choice([1, -1], size=N, p=[18/38, 20/38])
    return np.sum(money)
```

<h2>Task 2</h2>

<b>I used a histogram for plotting and manually adjusted the bin sizes to clearly visualize the normality of the distribution. As we can see, with an increase in the number of game rounds, the average decreases, and the standard deviation increases.
```python
iteration = 100000
N_list = [10, 25, 100, 1000]
money_10 = []
money_25 = []
money_100 = []
money_1000 = []
for i in N_list:
    y = []
    for _ in range(iteration):
        if(i == 10):
            money_10.append(simulate_N(i))
        elif(i == 25):
            money_25.append(simulate_N(i))
        elif(i == 100):
            money_100.append(simulate_N(i))
        else:
            money_1000.append(simulate_N(i))
money_10 = np.array(money_10)
money_25 = np.array(money_25)
money_100 = np.array(money_100)
money_1000 = np.array(money_1000)
```

```python
def theoretical_values(N):
    p_win = 18/38
    p_loss = 20/38
    expected_value = N * (p_win * 1 + p_loss * -1)
    variance = N * ((1 - expected_value/N)**2 * p_win + (-1 - expected_value/N)**2 * p_loss)
    standard_error = np.sqrt(variance)
    return expected_value, standard_error
theoretical_expected_10 , theoretical_SE_10 = theoretical_values(10)
theoretical_expected_25 , theoretical_SE_25 = theoretical_values(25)
theoretical_expected_100 , theoretical_SE_100 = theoretical_values(100)
theoretical_expected_1000 , theoretical_SE_1000 = theoretical_values(1000)
print("Expected value for playing 10 round:", theoretical_expected_10)
print("Standard error of money earned in 10 round:", theoretical_SE_10)
print("Expected value for playing 25 round:", theoretical_expected_25)
print("Standard error of money earned in 25 round:", theoretical_SE_25)
print("Expected value for playing 100 round:", theoretical_expected_100)
print("Standard error of money earned in 100 round:", theoretical_SE_100)
print("Expected value for playing 1000 round:", theoretical_expected_1000)
print("Standard error of money earned in 1000 round:", theoretical_SE_1000)
```

<h2>Task 5</h2>

<b>I assumed our player just play with casino</b>

```python
def casino_loose(N):
    p_win = 20/38
    p_loss = 18/38
    expected_value = N * (p_win * 1 + p_loss * -1)
    variance = N * ((1 - expected_value/N)**2 * p_win + (-1 - expected_value/N)**2 * p_loss)
    standard_error = np.sqrt(variance)
    return expected_value, standard_error
def clt_probability(N=25):
    mu, sigma = casino_loose(N)
    prob = norm.cdf(0, loc=mu, scale=sigma)
    return prob
```

```python
print(clt_probability(25))
```

```python
def monte_carlo_verification(N=25, simulations=100000):
    outcomes = np.random.choice([1, -1], size=(simulations, N), p=[20/38, 18/38])
    total_earnings = np.sum(outcomes, axis=1)
    probability = np.mean(total_earnings < 0)
    return probability
```

```python
print(monte_carlo_verification())
```
